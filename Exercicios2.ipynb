01. Faça um algoritmo em Python para ler dois valores numéricos e apresentar a diferença do maior
pelo menor.

```python
#Algoritmo de diferença:

x = float(input("Informe o primeiro valor:"))
y = float(input("Informe o segundo valor:"))

if x>y:
  x= x - y
  print("A diferença entre os números é:",round(x))
else:
  y = y - x
  print("A diferença entre os números é:",round(y))
```

mensagem: "NÚMEROS IGUAIS" e encerrar a execução; caso contrário, escrever o de maior
valor.

03. Faça um algoritmo em Python para ler dois valores: NUM1 e NUM2, e se NUM1 for maior que
NUM2 executa a soma de NUM1 e NUM2; caso contrário, executa uma subtração.

```python
# Variaveis
num1 = float(input("Digite o primeiro número:"))
num2 = float(input("Digite o segundo número:"))

if num1 > num2:
  print("A soma dos dois números é:",num1 + num2)
elif num1 == num2:
  print("Os números são iguais.")
else:
  print("A subtração dos dois números é:",num2 - num1)
```
